````python
'''
Rosalind_1 = 'ATCCAGCT'
Rosalind_2 = 'GGGCAACT'
Rosalind_3 = 'ATGGATCT'
Rosalind_4 = 'AAGCAACC'
Rosalind_5 = 'TTGGAACT'
Rosalind_6 = 'ATGCCATT'
Rosalind_7 = 'ATGGCACT'
'''

sequences = [
    'ATCCAGCT',
    'GGGCAACT',
    'ATGGATCT',
    'AAGCAACC',
    'TTGGAACT',
    'ATGCCATT',
    'ATGGCACT'
]

# 遍历每个序列 :比如你想打印每条序列的第一个碱基：
#for seq in sequences:
#    print(seq[0])

# 或者你想统计每个位置上 A/C/G/T 的数量：

# 初始化 profile 矩阵
profile = {
    'A': [0]*8,
    'C': [0]*8,
    'G': [0]*8,
    'T': [0]*8
}

# 遍历每条序列
for seq in sequences:
    for i, base in enumerate(seq):
        profile[base][i] += 1

# 打印 profile
for base in 'ACGT':
    print(f"{base}: {' '.join(map(str, profile[base]))}")

# 构建共识序列
consensus = ''
for i in range(8):
    max_base = max('ACGT', key=lambda b: profile[b][i])
    consensus += max_base

print("Consensus:", consensus)
````

🧠 目标：构建 profile 矩阵
我们有多条 DNA 序列，每条长度为 8。我们想统计每个位置上 A、C、G、T 出现的次数。

🔍 逐行解释
✅ 第一步：初始化 profile
```python
profile = {
    'A': [0]*8,
    'C': [0]*8,
    'G': [0]*8,
    'T': [0]*8
}
```
这行代码创建了一个字典 profile，它的结构是：

碱基|	列表含义
|---|---
|'A'	|长度为 8 的列表，表示 A 在每个位置出现的次数
|'C'	|同上
|'G'	|同上
|'T'	|同上

初始时，每个位置的计数都是 0。

✅ 第二步：遍历每条序列
```python
for seq in sequences:
```
sequences 是一个列表，包含多条 DNA 序列

每次循环，seq 是其中一条序列，比如 'ATCCAGCT'

✅ 第三步：遍历序列中的每个碱基
```python
for i, base in enumerate(seq):
```
enumerate(seq) 会把序列拆成 (索引, 字符) 对

例如：'ATCCAGCT' 会变成：

```代码
i=0, base='A'
i=1, base='T'
i=2, base='C'
i=3, base='C'
i=4, base='A'
i=5, base='G'
i=6, base='C'
i=7, base='T'
```

✅ 第四步：更新 profile 矩阵
```python
profile[base][i] += 1
```
这句是核心，意思是：

找到当前碱基 base（比如 'A'）

找到它在 profile 中对应的列表：profile['A']

找到它在第 i 个位置的计数：profile['A'][i]

把这个位置的计数加 1

🧪 举个例子
假设我们处理第一条序列 'ATCCAGCT'，我们会执行：

```python
profile['A'][0] += 1  # 第1位是A
profile['T'][1] += 1  # 第2位是T
profile['C'][2] += 1  # 第3位是C
profile['C'][3] += 1  # 第4位是C
profile['A'][4] += 1  # 第5位是A
profile['G'][5] += 1  # 第6位是G
profile['C'][6] += 1  # 第7位是C
profile['T'][7] += 1  # 第8位是T
```
这样就把第一条序列的碱基信息记录进 profile 了。

✅ 总结一句话
```python
for seq in sequences:
    for i, base in enumerate(seq):
        profile[base][i] += 1
```
这段代码的作用是：

遍历每条 DNA 序列的每个位置，把该位置的碱基计数加到 profile 矩阵中。



````python
# profile[base][i] += 1 的最原始版本：

# 初始化每个碱基的计数列表
A_counts = [0, 0, 0, 0, 0, 0, 0, 0]
C_counts = [0, 0, 0, 0, 0, 0, 0, 0]
G_counts = [0, 0, 0, 0, 0, 0, 0, 0]
T_counts = [0, 0, 0, 0, 0, 0, 0, 0]

# 所有 DNA 序列
sequences = [
    'ATCCAGCT',
    'GGGCAACT',
    'ATGGATCT',
    'AAGCAACC',
    'TTGGAACT',
    'ATGCCATT',
    'ATGGCACT'
]

# 遍历每条序列
for seq in sequences:
    for i in range(8):  # 每个位置
        base = seq[i]
        if base == 'A':
            A_counts[i] += 1
        elif base == 'C':
            C_counts[i] += 1
        elif base == 'G':
            G_counts[i] += 1
        elif base == 'T':
            T_counts[i] += 1
````
